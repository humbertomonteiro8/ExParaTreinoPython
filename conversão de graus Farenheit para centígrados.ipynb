1.12.3. A conversão de graus Farenheit para centígrados é obtida por C = - (F - 32).
Fazer um algoritmo que calcule e escreva uma tabela de graus centígrados em função de
graus Farenheit, que variam de 50 a 150 de 1 em 1.

```python
print('°F\t°C')

for f in range(50,150,1):
    c = round(5/9*(f-32),1)
    print(f,'\t',c)
  
```
